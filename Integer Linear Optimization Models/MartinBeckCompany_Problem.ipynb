Formulate a mixed-integer programming model that could be used to help
Martin-Beck determine which new plant or plants to open in order to satisfy
anticipated demand.

#### Sets and Indicies

$\mathcal{P}$: Plants <br>
$\mathcal{D}$: Distrubution Centers <br>

$p \in \mathcal{P}$: An index in the set $\mathcal{P}$, where $p_2, p_3, p_4, p_5$ are proposed plants  <br>
$d \in \mathcal{D}$: An index in the set $\mathcal{D}$ <br>

#### Data

$F$: A column vector representing Fixed cost for the Plants <br>
$f_p \in F, p \in \mathcal{P}$: An index representing Fixed cost for each the Plants <br>
$f_1$: Equals 0, since it is already built.

$C$: A column vector representing Capacity for the Plants <br>
$c_p \in C, p \in \mathcal{P}$: An index representing Capacity for each the Plants <br>

$M$: A column vector representing Demand for the Distribution Centers <br>
$m_d \in M, d \in \mathcal{D}$: An index representing Demand for the Distribution Centers <br>

$S$: A matrix represnting the Shipping Cost from each plant to each distribution center <br>
$s_{pd} \in S, p \in \mathcal{P}, d \in \mathcal{D}$: an Index in the set $S$ <br>
 
#### Decision Variables

$X$: A column vector representing which plants to build <br>
$x_p \in X, p \in \mathcal{P}$: An index in the set $X$ <br>

$Y$: A matrix representing the amount to ship between each distrubution center and plant <br>
$y_pd \in Y, p \in \mathcal{P}, d \in \mathcal{D}$: An index in the set $Y$ <br>


#### Function


\begin{align*}
\ {\mathrm{minimize }} & \; (F^T X) + 1_{||D||}^TYS1_{||D||}\\
\text{subject to:} \\
& Y1_{||D||} \leq C\\ 
& 1_{||P||}Y \leq M^T\\ 
& x_i \in \{0, 1\} ; \forall i\in\mathcal{P}\\
& y_{iu} \geq 0\ \text{and Integer}; \forall i\in\mathcal{P}, \forall u\in\mathcal{D}
\end{align*}
