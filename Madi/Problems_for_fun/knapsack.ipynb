```python
# i was curious how fast it can solve knapsack problem
from docplex.mp.model import Model
model = Model(name = 'knapsack')
```

```python
from random import randint
n = 100000
nums = [randint(100000, 100000000) for _ in range(n)]
weights = [randint(1, 100) for _ in range(n)]
cap = 300
```

```python
x = model.binary_var_list(n, name = 'x')
```

```python
model.maximize(model.sum(x[i] * nums[i] for i in range(n)))
```

```python
model.add_constraint(model.sum(x[i] * weights[i] for i in range(n)) <= cap)
```

```python
model.solve(log_output = True)
```

```python
obj = model.objective_value
assignment = [x[i].solution_value for i in range(n)]
print(obj, assignment)
```
