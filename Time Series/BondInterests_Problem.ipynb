Corporate triple A bond interest rates for 12 consecutive months
are as follows:

a. Construct a time series plot. What type of pattern exists in the data?

b. Develop three-month and four-month moving averages for this time series. Does the
three-month or the four-month moving average provide the better forecasts based on
MSE? Explain.

c. What is the moving average forecast for the next month?

```python
data = [9.5, 9.3, 9.4, 9.6, 9.8, 9.7, 9.8, 10.5, 9.9, 9.7, 9.6, 9.6]

import matplotlib.pyplot as plt
plt.ylim(bottom = 0, top = max(data)*1.5)
plt.plot(data)
```

```python
def calc_moving_avg(data, k):
    preds = []
    for i in range(k, len(data)+1):
        preds.append(sum(data[i-k:i]) / len(data[i-k:i]))

    return preds

four_month_avg = calc_moving_avg(data, 4)
three_month_avg = calc_moving_avg(data, 3)

mse = lambda X, Y: sum([(i-u)**2 for i, u in zip(X, Y)])/len(X)

print(mse(four_month_avg[:-1],  data[4:]))
print(mse(three_month_avg[:-1],  data[3:]))
```
