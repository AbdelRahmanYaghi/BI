```python
# 3 week moving average

def calc_moving_avg(data, k):
    preds = []
    for i in range(k, len(data)+1):
        preds.append(sum(data[i-k: i])/k)

    return preds

preds = calc_moving_avg(values, 3)
mse = sum([(i - u)**2 for i, u in zip(preds[:-1], values[3:])])/len(values[3:])
print(mse)
preds[-1]
```

```python
def calc_exponential_smoothing(data, alpha):
    preds = [data[0]]
    for i in range(1, len(data)):
        preds.append(alpha * data[i] + (1- alpha) * preds[i-1])

    return preds

preds = calc_exponential_smoothing(values, 0.3)

sum([(i - u)**2 for i, u in zip(preds[:-1], values[1:])]) / len(values[1:])
```
